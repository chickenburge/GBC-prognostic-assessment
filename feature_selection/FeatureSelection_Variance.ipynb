```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd
```

```python
dataFile = 'file_path'
data = pd.read_csv(dataFile)
X = data.iloc[: ,0:1502]
X.shape
```

```python
selector_var = VarianceThreshold(0.6)
X_var = selector_var.fit_transform(X)
X_var.shape
```

```python
selector_var.get_support()
```

```python
X.columns[selector_var.get_support()]
```

```python
X_var = pd.DataFrame(X_var, columns = X.columns[selector_var.get_support()])
X_var
```

```python
X_var.to_csv('file_path')
```
