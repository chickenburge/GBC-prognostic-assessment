## Get it started

### Obtain datasets 

````python
import tensorflow.compat.v1 as tf
import numpy as np
````

````python
import pandas as pd
from sklearn.model_selection import GridSearchCV
train_data = pd.read_csv("trainnm.csv")
test_data = pd.read_csv("testnm.csv")
train_data = train_data.iloc[: ,0:25]
test_data = test_data.iloc[: ,0:25]
train_data.head()
````

### Dataset statistics

````python
from tfdeepsurv.datasets import survival_stats

# specify the colnames of observed status and time in your dataset
colname_e = 'event'
colname_t = 'time'

survival_stats(train_data, t_col=colname_t, e_col=colname_e, plot=True)
````

````python
survival_stats(test_data, t_col=colname_t, e_col=colname_e, plot=True)
````

### Survival data transfrom

The transformed survival data includes the existing covariates and a new label column. In the new label column, a negative value indicates that this one is a right-censored sample, and a positive value indicates an event occurrence. The new label column 'Y' is simply generated from the time and event columns according to the below equation. 

```
Y =  time, if event = 1
Y = -time, if event = 0
```

**NOTE**: In the latest version 2.0, survival data must be transformed by the functionality `tfdeepsurv.datasets.survival_df`.

````python
from tfdeepsurv.datasets import survival_df

surv_train = survival_df(train_data, t_col=colname_t, e_col=colname_e, label_col="Y")
surv_test = survival_df(test_data, t_col=colname_t, e_col=colname_e, label_col="Y")

# columns 't' and 'e' are packed into an new column 'Y'
surv_train.head()
````

### Model initialization

**NOTE:** You can freely change all hyper-parameters during model initialization or training as you want.

All hyper-parameters is as follows:
- `nn_config`: model parameters configuration
- `hidden_layers_nodes`: hidden layers configuration
- `num_steps`: training steps

Hyperparameters tuning can refer to README in directory `byopt` of this project.

````python
from tfdeepsurv import dsnn

# Number of features in your dataset
input_nodes = len(surv_train.columns) - 1
# Specify your neural network structure
hidden_layers_nodes = [12, 6, 2, 1]

# the arguments of dsnn can be obtained by Bayesian Hyperparameters Tuning.
# It would affect your model performance largely!
nn_config = {
    "learning_rate": 0.07,#0.001
    "learning_rate_decay": 0.7,#2.922e-04
    "activation": 'tanh', #['relu', 'tanh', 'sigmoid']
    "L1_reg": 3.4e-4, 
    "L2_reg": 8.8e-4, 
    "optimizer": 'adam',#['sgd', 'adam', 'rms']
    "dropout_keep_prob": 1,
    "seed": 2
}
# ESSENTIAL STEP-1: Pass arguments
model = dsnn(
    input_nodes, 
    hidden_layers_nodes,
    nn_config
)

# ESSENTIAL STEP-2: Build Computation Graph
model.build_graph()
````

### Model training

You can save your trained model by passing `save_model="file_name.ckpt"` or load your trained model by passing `load_model="file_name.ckpt"`

````python
Y_col = ["Y"]
X_cols = [c for c in surv_train.columns if c not in Y_col]

# model saving and loading is also supported!
# read comments of `train()` function if necessary.

# ESSENTIAL STEP-3: Train Model
# `num_steps` is also an important parameters
watch_list = model.train(
    surv_train[X_cols], surv_train[Y_col],
    num_steps=2500,
    num_skip_steps=100,
    plot=True
)
````

### Model evaluation

````python
print("CI on training data:", model.evals(surv_train[X_cols], surv_train[Y_col]))
print("CI on test data:", model.evals(surv_test[X_cols], surv_test[Y_col]))
````

````python
def bootstrap_ci(features, labels, model, n_iterations=100, random_seed=42):
    np.random.seed(random_seed)  # random seed
    ci_scores = []
    n_samples = len(features)
    for _ in range(n_iterations):
        # Perform resampling with replacement using DataFrame indexing
        indices = np.random.choice(features.index, size=n_samples, replace=True)
        resampled_features = features.iloc[indices]
        resampled_labels = labels.iloc[indices]
        # Calculate the C-index for the resampled data
        ci = model.evals(resampled_features, resampled_labels)
        ci_scores.append(ci)
    return np.array(ci_scores)
# 获取训练数据的特征和标签 Features a
train_features_cindex = surv_train[X_cols]
train_labels_cindex = surv_train[Y_col]
# 获取测试数据的特征和标签
test_features_cindex = surv_test[X_cols]
test_labels_cindex = surv_test[Y_col]
# 计算训练数据的C-index
train_ci_scores = bootstrap_ci(train_features_cindex, train_labels_cindex, model)
# 计算测试数据的C-index
test_ci_scores = bootstrap_ci(test_features_cindex, test_labels_cindex, model)

# 计算训练数据的均值、标准差和95%置信区间
z_value = 1.96
train_ci_mean = np.mean(train_ci_scores)
train_ci_std = np.std(train_ci_scores)
train_ci_low = train_ci_mean - z_value * train_ci_std
train_ci_up = train_ci_mean + z_value * train_ci_std
train_ci_confidence_interval = np.percentile(train_ci_scores, [2.5, 97.5])
# 计算测试数据的均值、标准差和95%置信区间
test_ci_mean = np.mean(test_ci_scores)
test_ci_std = np.std(test_ci_scores)
test_ci_low = test_ci_mean - z_value * test_ci_std
test_ci_up = test_ci_mean + z_value * test_ci_std
test_ci_confidence_interval = np.percentile(test_ci_scores, [2.5, 97.5])
# 打印结果
print(f"训练数据自助法C-index：均值 = {train_ci_mean}, 标准差 = {train_ci_std}")
print(f"训练数据自助法C-index 95% 置信区间：{train_ci_low}, {train_ci_up}")
print(f"测试数据自助法C-index：均值 = {test_ci_mean}, 标准差 = {test_ci_std}")
print(f"测试数据自助法C-index 95% 置信区间：{test_ci_low}, {test_ci_up}")
````

````python
# 交叉验证
all_data=train_data
from sklearn.model_selection import KFold
n_folds = 5
kf=KFold(n_splits=n_folds, shuffle=True, 
         random_state=1)
scoreTrainList, scoreValList = [], []

for train_index, val_index in kf.split(all_data):
    train_sett = all_data.iloc[train_index]
    val_sett = all_data.iloc[val_index]
    from tfdeepsurv import dsnn
    input_nodes = 23#len(train_data.columns) - 1
    hidden_layers_nodes = [12, 6, 2, 1]
    nn_config = {
    "learning_rate": 0.07,#0.001
    "learning_rate_decay": 0.7,#2.922e-04
    "activation": 'tanh', #['relu', 'tanh', 'sigmoid']
    "L1_reg": 3.4e-4, 
    "L2_reg": 8.8e-4, 
    "optimizer": 'adam',#['sgd', 'adam', 'rms']
    "dropout_keep_prob": 1,
    "seed": 2
}

    model1 = dsnn(
        input_nodes, 
        hidden_layers_nodes,
        nn_config
    )

    model1.build_graph()
    Y_col = ["Y"]
    X_cols = [c for c in surv_train.columns if c not in Y_col]
    watch_list = model1.train(
        train_sett[X_cols], train_sett[Y_col],
        num_steps=2500,
        num_skip_steps=100,
        plot=True
    )
    scoreTrainList.append(model1.evals(train_sett[X_cols], train_sett[Y_col]))
    scoreValList.append(model1.evals(surv_test[X_cols], surv_test[Y_col]))
print(scoreTrainList)
print(scoreValList)
````

````python
print(np.mean(scoreTrainList))
print(np.std(scoreTrainList))
print(np.mean(scoreValList))
print(np.std(scoreValList))
````
